```python
cipher1 = "dyvkrwwrzbcksvz:xkdyeck@bv@rbvukr:ukw/btvckz:kbvruz:vccks/dykgzdy/edkr:ukgzdyz:kdyvkbvr,;mkdyv:kcyr,,kzdksvkdz;vkd/kcvdkdyvkczhkxv:d,v;v:kd/kg/b.nkdr.z:xk/buvbmke@/:kdyvkrtt/;@,zcyz:xk/wkdyvzbkuvczx:mkzk;riksvkceuuv:,ikdbr:c@/bdvuk/edk/wkdyzck@,rtvmkr:ukdyrdkr,,ki/ebkw/btvckz:kdyvkcr;vkdz;vksvk/:kdyvkwzv,ukd/k;vvdk;vl"

chars = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z', ' ', '.', ',', ';', ':', '/', '@']

freq = {}
pos = {}

for i,c in enumerate(chars):
    freq[c] = 0
    pos[c] = i

for c in cipher1:
    freq[c] += 1

arr = []

for ch, count in freq.items():
    arr.append([count, ch])

arr.sort( reverse=True)

for [count, letter] in arr:
    print(f"{letter}: {count}")
```

From the above frequency distribution, seems like k is space, and the cipher is shifted by 16 characters.

```python
def decypher(cypher):
    res = ""
    for letter in cypher:
        idx = (pos[letter]+16)%len(chars)
        res += chars[idx]
    return res

print(decypher(cipher1))
```
